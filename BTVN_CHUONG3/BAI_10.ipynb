```python
import pandas as pd

data = {
    'continent': ['Europe', 'Asia', 'Africa', 'Europe', 'North America', 'South America', 'Oceania', 'Europe', 'Asia', 'Africa'],
    'country': ['Germany', 'China', 'Nigeria', 'France', 'USA', 'Argentina', 'Australia', 'Spain', 'India', 'South Africa'],
    'beer_servings': [200, 50, 10, 150, 250, 100, 220, 180, 30, 40],
    'wine_servings': [150, 10, 5, 180, 50, 60, 40, 130, 5, 20],
    'spirit_servings': [100, 40, 20, 120, 130, 90, 150, 110, 30, 80]
}

df = pd.DataFrame(data)

df.to_csv('drinks.csv', index=False)

print("Dữ liệu đã được tạo và lưu vào drinks.csv:")
print(df)
```

```python
import pandas as pd

drink = pd.read_csv('drinks.csv', index_col=0)

print("Kiểu dữ liệu của drink:", type(drink))
print("Kích thước của drink:", drink.shape)

print("\nTên các cột trong drink:")
print(drink.columns)

print("\n5 dòng đầu tiên:")
print(drink.head())

print("\n5 dòng cuối cùng:")
print(drink.tail())

avg_beer_consumption_by_continent = drink.groupby('continent')['beer_servings'].mean()
print("\nSố lượng bia tiêu thụ trung bình ở mỗi châu lục:")
print(avg_beer_consumption_by_continent)

wine_servings_by_continent = drink.groupby('continent')['wine_servings'].describe()
print("\nThông tin thống kê số lượng rượu vang tiêu thụ ở mỗi châu lục:")
print(wine_servings_by_continent)

avg_consumption_by_continent = drink.groupby('continent')[['beer_servings', 'wine_servings']].mean()
print("\nSố lượng bia và rượu tiêu thụ trung bình ở mỗi châu lục:")
print(avg_consumption_by_continent)

median_consumption_by_continent = drink.groupby('continent')[['beer_servings', 'wine_servings']].median()
print("\nGiá trị trung vị cho các loại bia và rượu tiêu thụ ở mỗi châu lục:")
print(median_consumption_by_continent)

spirit_servings_by_continent = drink.groupby('continent')['spirit_servings'].agg(['mean', 'max', 'min'])
print("\nNượu mạnh tiêu thụ trung bình, lớn nhất và nhỏ nhất ở mỗi châu lục:")
print(spirit_servings_by_continent)

sorted_by_beer = drink.sort_values(by='beer_servings', ascending=True)

top_5_beer = sorted_by_beer.tail(5)
print("\n5 quốc gia tiêu thụ bia nhiều nhất:")
print(top_5_beer[['beer_servings']])

bottom_5_beer = sorted_by_beer.head(5)
print("\n5 quốc gia tiêu thụ bia ít nhất:")
print(bottom_5_beer[['beer_servings']])
```
